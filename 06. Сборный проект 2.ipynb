### Проверка эффективности обогащения.

```python
def recovery(df):
    c = df['rougher.output.concentrate_au']
    f = df['rougher.input.feed_au']
    t = df['rougher.output.tail_au']
    
    return (c * (f - t)) / (f * (c - t)) * 100
```

```python
target_recovery = gold_recovery_train['rougher.output.recovery']
calculate_recovery = recovery(gold_recovery_train)

mean_absolute_error(target_recovery, calculate_recovery)
```

#### Вывод.

Рассчитанное значение MAE близко к нулю, следовательно, эффективность обогащения вычислена правильно.

```python
def unique_stages(data=gold_recovery_full):
    columns = data.columns
    stages = []
    
    for column in columns:
        val = column.split('.')[0]
        
        if val not in stages:
            stages.append(val)
            
    return stages
```

```python
def interest_feature(feature, data=gold_recovery_full, step='output'):
    stages = unique_stages(data)
    vals = {}
    
    for stage in stages:
        try:
            vals[stage] = data['.'.join([stage, step, feature])]
        
        except:
            continue
        
    return pd.DataFrame(vals)
```

### Суммарная концентрация веществ на разных стадиях.

```python
def summary_concentrate(stage_name=None, data=gold_recovery_full):
    metal = ['ag', 'pb', 'au', 'sol']
    
    if stage_name == 'input':
        stage = 'rougher.input.feed_'
        
    elif stage_name == 'output':
        stage = 'final.output.concentrate_'
        
    else:
        stage = 'rougher.output.concentrate_'
        
    temp = {}
    
    for column in data.columns:
        for me in metal:
            substring = stage + me
            
            try:
                if substring in column:
                    temp[column] = data[column]

            except:
                continue
                
    name = '.'.join(substring.split('.')[:-1])
            
    return name, pd.DataFrame(temp).sum(axis=1)
```

```python
def calc_conc(data):
    summary_conc = {}

    for stage in ('input', 'output', None):
        name, val = summary_concentrate(stage_name=stage, data=data)
        summary_conc[name] = val

    summary_conc = pd.DataFrame(summary_conc)
    
    return summary_conc
```

## Модель

### Функция для вычисления sMAPE.

```python
def smape(target, predict):
    smape = abs(target - predict)
    smape /= (abs(target) + abs(predict)) / 2
    smape *= 100
    smape = sum(smape) / len(target)
    
    return smape
```

```python
def summary_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

### Модели без подбора оптимальных параметров.

```python
def model_no_params(model):
    smape_score = {}

    for column in target_columns:
        scores = cross_val_score(model,
                                 features_train,
                                 target_train[column],
                                 scoring=smape_scorer,
                                 cv=5)

        final_score = scores.mean()

        smape_score[column.split('.')[0]] = abs(final_score)

        print(f'Средняя оценка sMAPE модели для {column}:', abs(final_score))
        
    return smape_score
```
